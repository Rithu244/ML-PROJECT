**What is Clustering?**

Clustering is a task of grouping a set of objects in such a way that objects in the same group are more similar to each other than to those in other group. K-means clustering is a widely used method for cluster analysis

**For Example:-**
Imagine that you are a Data Scientist working for a retail company and your boss requests for the customersâ€™ segmentation into the following groups: low, average, medium, or platinum customers based on spending behavior for targeted marketing purposes and product recommendations. This helps us to make better marketing strategy

Knowing that there is no such historical label associated with those customers, how is it possible to categorize them?

This is where clustering can help. It is an unsupervised machine-learning technique used to group unlabeled data into similar categories or clusters.

## What is K-Means Clustering?
The idea behind K-means clustering is to divide a dataset into a specified number of clusters (k), where all the points within the same cluster are similar to one another, and those in different clusters are different.

It starts by randomly assigning each data point to a cluster, and then it iteratively improves the clusters by moving the data points to the cluster center that is closest to them. This logic continues until the cluster assignments stop changing, or a maximum number of iterations is reached.

### What are the key steps of K-means clustering?
Below are the five main steps of the k-means algorithm:

1. Specify the Number of cluster 'k'
2. Randomly initialize the seed or (cluster centers or centroids)
3. Assign each data point to the closest cluster center
4. Recompute the cluster's center as the mean of all data in the cluster.
5. Repeat steps 3 and 4 until the cluster assignment stop changing/maximum iteration is reached

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Create the dataset
dataset = pd.DataFrame({'X': [2, 2, 5, 4, 8, 6, 5, 5, 6, 4], 'y': [4, 6, 6, 7, 3, 6, 2, 7, 3, 4]})

# Initialize the KMeans model with k=3
kmeans = KMeans(n_clusters=3)

# Fit the model to the data
kmeans.fit(dataset)

# Add the cluster labels to the DataFrame
dataset['Cluster'] = kmeans.labels_

# Print the DataFrame with cluster labels
print(dataset)

# Plot the data points with cluster centers
plt.scatter(dataset['X'], dataset['y'], c=dataset['Cluster'], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='black', marker='x', s=200)
plt.xlabel('X')
plt.ylabel('y')
plt.title('K-Means Clustering (k=3)')
plt.show()
```

### Choosing the optimum number of clusters

### Determine the best number of cluster
A clustering model will not be relevant if we fail to identify the correct number of clusters to consider. Multiple techniques exist in the literature. We are going to consider the Elbow method, one of the widely used to find the optimal number of clusters.

The first helper function creates for each value of K the corresponding KMeans model and saves its inertia along with the actual K value.

The second function uses those inertias and K values to generate the final Elbow plot.

### **WCSS** --> [Within Cluster Sum of Squares]

```python
# Now will find the wcss value for different number of cluster

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(dataset[['X','y']])

    wcss.append(kmeans.inertia_)
```

```python
# Now will plot the elbow graph to select optimal value of cluster

plt.plot(range(1,11),wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()
```

#### So we can see there is a sudden drop when the No of cluster is 3, and after that there is no significant drop, so we are considering the optimal value of cluster(k) = 3

### silhouette score

The silhouette score is a measure of the quality of the clusters formed by a clustering algorithm, It explains how well the data is clusters and separated, The silhouette score ranges from -1 to 1, with higher values indicating better cluster separation and cohesion


```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(dataset[['X', 'y']], dataset['Cluster'].values)
print(f"Silhouette Score (Using scikit-learn): {silhouette_avg}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(dataset[['X', 'y']])

    # Check the number of unique labels
    unique_labels = np.unique(labels)
    if len(unique_labels) > 1 and len(unique_labels) < len(dataset):
        silhouette_avg = silhouette_score(dataset[['X', 'y']], labels, metric='euclidean')
        silhouette_scores.append(silhouette_avg)
    else:
        # Handle the case where the number of unique labels is not valid
        silhouette_scores.append(None)

plt.figure(figsize=(14, 7))
plt.plot(range(2, 11), silhouette_scores, marker='*')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.show()
```

#### Based on the silhouette scores calculated for different values of k (number of clusters), it has been observed that the optimal number of clusters is 3
