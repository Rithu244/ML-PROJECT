## Hierarchical clustering

Hierarchical clustering is a method for cluster analysis that builds a hierarchy of clusters by either successively merging or dividing data points. There are two main types of hierarchical clustering: agglomerative (bottom-up) and divisive (top-down). Agglomerative clustering is the more common approach, where you start with individual data points as separate clusters and then merge them step by step.

### Here are the steps to perform hierarchical clustering:

1. Start with each data point as its own cluster.
2. Calculate the distance (similarity) between each pair of clusters.
3. Merge the two clusters with the smallest distance into a new cluster.
4. Repeat steps 2 and 3 until there is only one cluster left, which contains all data points.

```python
import pandas as pd
```

```python
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Create a DataFrame from your data
df = pd.DataFrame({'Food Exp':[2,8,9,1,8.5],'Cloth Exp':[4,2,3,5,1]})

# Perform hierarchical clustering with single linkage
linkage_matrix = linkage(df, method='single')  # Use 'single' for single linkage

# Plot the dendrogram to visualize the clustering
plt.figure(figsize=(8, 5))
dendrogram(linkage_matrix, labels=['a','b','c','d','e'])
plt.title("Hierarchical Clustering Dendrogram (Single Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```
